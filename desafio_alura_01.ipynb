```python
import pandas as pd
import glob

# Passo 2: Listar os arquivos CSV na pasta local
file_path = '/content/*.csv'
csv_files = glob.glob(file_path)

# Lista para armazenar cada DataFrame individual
dataframes = []

# Passo 3 e 4: Ler cada arquivo CSV e adicionar à lista de DataFrames
for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)

# Passo 5: Concatenar todos os DataFrames em um único DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

# Exibir o DataFrame combinado
print(combined_df)
```

```python
# Passo 1: Eliminar as linhas com dados nulos
df_sem_nulos = combined_df.dropna()

# Passo 2: Remover as linhas duplicadas
emprestimos_df = df_sem_nulos.drop_duplicates()

# Exibir o DataFrame final
print(emprestimos_df)
```

```python
import pandas as pd

# URL do arquivo Parquet
url = "https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet"

# Ler o arquivo Parquet diretamente da URL
exemplares_df = pd.read_parquet(url, engine='pyarrow')

# Eliminar as linhas com dados nulos
exemplares_sem_nulos = exemplares_df.dropna()

# Remover as linhas duplicadas
exemplares_final = exemplares_sem_nulos.drop_duplicates()

# Exibir o DataFrame final
print(exemplares_final)
```

```python
exemplares_final.head()
```

```python
import pandas as pd

# Exemplo de criação dos DataFrames para ilustração (remova esta parte se já tem os DataFrames prontos)
# emprestimos_df = combined_df.copy()  # Supondo que combined_df é o emprestimos_df
# exemplares_df = exemplares_final.copy()  # Supondo que exemplares_final é o exemplares_df

# Mesclar os DataFrames através do campo codigo_barras
dados_acervo = pd.merge(
    emprestimos_df,
    exemplares_final,
    on='codigo_barras',
    how='inner'
)

# Exibir o DataFrame mesclado
print(dados_acervo)
```
